```python
try:
    euro12 = pd.read_csv('euro12.csv')
except FileNotFoundError:
    data = {
        'Team': ['Spain', 'Germany', 'Portugal', 'Italy', 'England'],
        'Goals': [12, 10, 6, 6, 5],
        'Yellow Cards': [11, 15, 12, 16, 5],
        'Red Cards': [0, 0, 0, 1, 0],
        'Shooting Accuracy': ['43%', '40%', '32%', '33%', '35%']
    }
    euro12 = pd.DataFrame(data)

print("\n--- Bài 4.20 (Euro2012) ---")
print("1. Goals:\n", euro12['Goals'])
print("2. Số đội tham gia:", euro12.shape[0])
print("3. Thông tin DataFrame:")
euro12.info()

discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print("\n4. DataFrame Discipline:\n", discipline)

discipline_sorted = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=False)
print("\n5. Discipline sắp xếp:\n", discipline_sorted)

avg_yellow_cards = euro12['Yellow Cards'].mean()
print("\n6a. Trung bình Yellow Cards:", avg_yellow_cards)

teams_gt_6_goals = euro12[euro12['Goals'] > 6]
print("\n6b. Đội ghi > 6 bàn:\n", teams_gt_6_goals)

teams_start_G = euro12[euro12['Team'].str.startswith('G')]
print("\n7. Đội bắt đầu bằng 'G':\n", teams_start_G)

print("\n8. 7 cột đầu:\n", euro12.iloc[:, :7].head())

print("\n9. Tất cả trừ 3 cột cuối:\n", euro12.iloc[:, :-3].head())

selected_cols = euro12[['Team', 'Goals', 'Shooting Accuracy', 'Yellow Cards', 'Red Cards']]
print("\n10. Các cột được chọn:\n", selected_cols.head())


teams_list = ['England', 'Italy', 'Russia']
euro_subset = euro12[euro12['Team'].isin(teams_list)]

result_11 = euro_subset[['Team', 'Shooting Accuracy']]
print("\n11. Subset:\n", result_11)
```

```python
try:
    drinks = pd.read_csv('drinks.csv', index_col=0)
except FileNotFoundError:
    
    data = {
        'beer_servings': [0, 89, 219, 107, 167],
        'spirit_servings': [0, 132, 110, 7, 263],
        'wine_servings': [0, 49, 0, 86, 171],
        'total_litres_of_pure_alcohol': [0.0, 4.9, 5.7, 5.0, 8.4],
        'continent': ['AS', 'EU', 'AS', 'EU', 'AF']
    }
    drinks = pd.DataFrame(data, index=['Afghanistan', 'Albania', 'Algeria', 'Andorra', 'Angola'])

print("\n--- Bài 4.21 (Drinks) ---")
print("1. Kiểu dữ liệu và cột:")
drinks.info()
print("\nKích thước (shape):", drinks.shape)
print("Các cột:", drinks.columns.tolist())

print("\n2. 5 dòng đầu:\n", drinks.head())
print("\n5 dòng cuối:\n", drinks.tail())

avg_beer_by_continent = drinks.groupby('continent')['beer_servings'].mean()
print("\n3. Trung bình bia tiêu thụ theo châu lục:\n", avg_beer_by_continent)

print("\n4. Thống kê rượu vang:\n", drinks['wine_servings'].describe())

avg_alcohol_by_continent = drinks.groupby('continent')[['beer_servings', 'wine_servings']].mean()
print("\n5. Trung bình bia và rượu theo châu lục:\n", avg_alcohol_by_continent)

median_max_min_spirit = drinks.groupby('continent')['spirit_servings'].agg(['median', 'max', 'min'])
print("\n6. Trung vị/Min/Max rượu mạnh theo châu lục:\n", median_max_min_spirit)

drinks_sorted = drinks.sort_values(by='beer_servings')
print("\n7. Sắp xếp dữ liệu (tăng dần theo bia):\n", drinks_sorted[['beer_servings']].head())

most_beer = drinks.sort_values(by='beer_servings', ascending=False).head(5)
print("\n5 quốc gia tiêu thụ bia nhiều nhất:\n", most_beer['beer_servings'])

least_beer = drinks.sort_values(by='beer_servings', ascending=True).head(5)
print("\n5 quốc gia tiêu thụ bia ít nhất:\n", least_beer['beer_servings'])
```

```python
stocks1_data = {
    'date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02', '2023-01-03'],
    'symbol': ['AAPL', 'GOOG', 'AAPL', 'GOOG', 'AAPL'],
    'open': [170.0, 95.0, 172.0, 96.0, 175.0],
    'high': [172.5, np.nan, 174.0, 98.0, 176.0],
    'low': [169.0, 94.0, 171.0, np.nan, 174.5],
    'close': [172.0, 96.0, 173.5, 97.5, 175.5],
    'volume': [100000, 50000, 120000, 60000, 110000]
}
stocks1 = pd.DataFrame(stocks1_data)

stocks2_data = {
    'date': ['2023-01-03', '2023-01-04', '2023-01-04'],
    'symbol': ['GOOG', 'AAPL', 'GOOG'],
    'open': [97.0, 176.0, 98.0],
    'high': [98.5, 177.0, 99.0],
    'low': [96.5, 175.0, 97.0],
    'close': [98.0, 176.5, 98.5],
    'volume': [55000, 130000, 65000]
}
stocks2 = pd.DataFrame(stocks2_data)

companies_data = {
    'symbol': ['AAPL', 'GOOG', 'MSFT'],
    'name': ['Apple Inc.', 'Alphabet Inc.', 'Microsoft Corp.'],
    'employees': [164000, 186770, 221000],
    'headquarters_city': ['Cupertino', 'Mountain View', 'Redmond'],
    'headquarters_state': ['CA', 'CA', 'WA']
}
companies = pd.DataFrame(companies_data)

print("\n--- Bài 4.22 (Stocks & Companies - Dữ liệu Mẫu) ---")

print("\n1a. stocks1.csv (5 dòng đầu):\n", stocks1.head())
print("1b. stocks2.csv (5 dòng đầu):\n", stocks2.head())
print("1c. companies.csv (5 dòng đầu):\n", companies.head())

print("\n2. Xử lý Null trong stocks1:")
print("Tổng số Null trước khi xử lý:\n", stocks1.isnull().sum())

stocks1['high'] = stocks1.groupby('symbol')['high'].transform(lambda x: x.fillna(x.max()))
stocks1['low'] = stocks1.groupby('symbol')['low'].transform(lambda x: x.fillna(x.min()))

print("\nTổng số Null sau khi xử lý:\n", stocks1.isnull().sum())
print("stocks1 sau xử lý Null (high/low):\n", stocks1)


stocks_merged = pd.concat([stocks1, stocks2], ignore_index=True)
print("\n3. stocks_merged (15 dòng cuối):\n", stocks_merged.tail(15))

stocks_companies = pd.merge(stocks_merged, companies, on='symbol', how='left')
print("\n4. stocks_companies (5 dòng đầu):\n", stocks_companies.head())

avg_price_volume = stocks_companies.groupby('symbol')[['open', 'high', 'low', 'close', 'volume']].mean()
print("\n5. Trung bình giá/volume theo công ty:\n", avg_price_volume)

close_stats = stocks_companies.groupby('symbol')['close'].agg(['mean', 'max', 'min'])
print("\n6. Thống kê giá đóng cửa theo công ty:\n", close_stats)

stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])
print("\n7. stocks_companies sau parsed_time (5 dòng đầu):\n", stocks_companies[['date', 'parsed_time']].head())

stocks_companies['result'] = np.where(stocks_companies['close'] > stocks_companies['open'], 'up', 'down')
print("\n8. Thêm cột 'result' (5 dòng đầu):\n", stocks_companies[['open', 'close', 'result']].head())
```
