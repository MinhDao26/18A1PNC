```python
import sqlite3

def init_db():
    conn = sqlite3.connect('product.db')
    c = conn.cursor()
    c.execute('''CREATE TABLE IF NOT EXISTS product
                 (Id INTEGER PRIMARY KEY, Name TEXT NOT NULL, 
                  Price REAL NOT NULL, Amount INTEGER NOT NULL)''')
    conn.commit()
    conn.close()

def add_product():
    id_prod = int(input("Nhập ID: "))
    name = input("Nhập tên sản phẩm: ")
    price = float(input("Nhập giá: "))
    amount = int(input("Nhập số lượng: "))
    
    conn = sqlite3.connect('product.db')
    c = conn.cursor()
    try:
        c.execute("INSERT INTO product VALUES (?, ?, ?, ?)", (id_prod, name, price, amount))
        conn.commit()
        print("Thêm thành công!")
    except sqlite3.IntegrityError:
        print("Lỗi: ID đã tồn tại.")
    conn.close()

def show_products():
    conn = sqlite3.connect('product.db')
    c = conn.cursor()
    c.execute("SELECT * FROM product")
    rows = c.fetchall()
    print("\n--- DANH SÁCH SẢN PHẨM ---")
    print(f"{'ID':<5} {'Tên':<20} {'Giá':<10} {'Số lượng':<10}")
    for row in rows:
        print(f"{row[0]:<5} {row[1]:<20} {row[2]:<10} {row[3]:<10}")
    conn.close()

def search_product():
    name = input("Nhập tên cần tìm: ")
    conn = sqlite3.connect('product.db')
    c = conn.cursor()
    c.execute("SELECT * FROM product WHERE Name LIKE ?", ('%' + name + '%',))
    rows = c.fetchall()
    if rows:
        for row in rows:
            print(row)
    else:
        print("Không tìm thấy.")
    conn.close()

def update_product():
    id_prod = int(input("Nhập ID sản phẩm cần sửa: "))
    new_price = float(input("Nhập giá mới: "))
    new_amount = int(input("Nhập số lượng mới: "))
    
    conn = sqlite3.connect('product.db')
    c = conn.cursor()
    c.execute("UPDATE product SET Price = ?, Amount = ? WHERE Id = ?", (new_price, new_amount, id_prod))
    if c.rowcount > 0:
        print("Cập nhật thành công.")
    else:
        print("Không tìm thấy ID này.")
    conn.commit()
    conn.close()

def delete_product():
    id_prod = int(input("Nhập ID sản phẩm cần xóa: "))
    conn = sqlite3.connect('product.db')
    c = conn.cursor()
    c.execute("DELETE FROM product WHERE Id = ?", (id_prod,))
    if c.rowcount > 0:
        print("Xóa thành công.")
    else:
        print("Không tìm thấy ID này.")
    conn.commit()
    conn.close()

init_db()
while True:
    print("\n1. Hiển thị danh sách")
    print("2. Thêm sản phẩm")
    print("3. Tìm kiếm sản phẩm")
    print("4. Cập nhật (Giá/Số lượng)")
    print("5. Xóa sản phẩm")
    print("0. Thoát")
    choice = input("Chọn chức năng: ")
    
    if choice == '1': show_products()
    elif choice == '2': add_product()
    elif choice == '3': search_product()
    elif choice == '4': update_product()
    elif choice == '5': delete_product()
    elif choice == '0': break
    else: print("Chọn sai, vui lòng chọn lại.")
```

```python
import sqlite3

def setup_employee_db():
    conn = sqlite3.connect('ql_nhan_vien.db')
    c = conn.cursor()
    c.execute('''
        CREATE TABLE IF NOT EXISTS PHONG (
            Id INTEGER PRIMARY KEY,
            Name TEXT NOT NULL
        )
    ''')
    c.execute('''
        CREATE TABLE IF NOT EXISTS NHAN_VIEN (
            id INTEGER PRIMARY KEY,
            ho_ten TEXT NOT NULL,
            tuoi INTEGER NOT NULL,
            dia_chi TEXT NOT NULL,
            luong REAL NOT NULL,
            Id_phong INTEGER,
            FOREIGN KEY (Id_phong) REFERENCES PHONG(Id)
        )
    ''')
    c.execute("SELECT COUNT(*) FROM PHONG")
    if c.fetchone()[0] == 0:
        ds_phong = [(1, 'Phòng Kế toán'), (2, 'Phòng IT'), (3, 'Phòng Nhân sự')]
        c.executemany("INSERT INTO PHONG (Id, Name) VALUES (?, ?)", ds_phong)
        print("Đã khởi tạo dữ liệu phòng ban.")

    conn.commit()
    conn.close()
    print("Cấu trúc CSDL Nhân viên đã sẵn sàng.")

def add_employee():
    print("\n--- Thêm Nhân Viên ---")
    e_id = int(input("Nhập ID nhân viên: "))
    name = input("Họ tên: ")
    age = int(input("Tuổi: "))
    addr = input("Địa chỉ: ")
    salary = float(input("Lương: "))
    dept_id = int(input("ID Phòng ban (1-Kế toán, 2-IT, 3-Nhân sự): "))
    
    conn = sqlite3.connect('ql_nhan_vien.db')
    c = conn.cursor()
    try:
        c.execute('''INSERT INTO NHAN_VIEN (id, ho_ten, tuoi, dia_chi, luong, Id_phong) 
                     VALUES (?, ?, ?, ?, ?, ?)''', 
                     (e_id, name, age, addr, salary, dept_id))
        conn.commit()
        print("Thêm nhân viên thành công!")
    except sqlite3.IntegrityError as e:
        print(f"Lỗi: {e} (Có thể trùng ID hoặc Phòng ban không tồn tại)")
    finally:
        conn.close()

def list_employees():
    conn = sqlite3.connect('ql_nhan_vien.db')
    c = conn.cursor()
    query = '''
        SELECT nv.id, nv.ho_ten, nv.tuoi, nv.luong, p.Name 
        FROM NHAN_VIEN nv
        LEFT JOIN PHONG p ON nv.Id_phong = p.Id
    '''
    c.execute(query)
    rows = c.fetchall()
    
    print("\n--- DANH SÁCH NHÂN VIÊN ---")
    print(f"{'ID':<5} {'Họ tên':<20} {'Tuổi':<5} {'Lương':<10} {'Phòng ban'}")
    for row in rows:
        dept_name = row[4] if row[4] else "N/A"
        print(f"{row[0]:<5} {row[1]:<20} {row[2]:<5} {row[3]:<10} {dept_name}")
    conn.close()
setup_employee_db()
while True:
    print("\n1. Thêm nhân viên")
    print("2. Xem danh sách nhân viên")
    print("0. Thoát")
    c = input("Chọn: ")
    if c == '1': add_employee()
    elif c == '2': list_employees()
    elif c == '0': break
```
