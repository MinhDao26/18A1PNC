```python
import threading
import time
import random
import requests 
import math
```

```python
#6.1
def worker_6_1():
    thread_name = threading.current_thread().name
    print(f"Xin chào từ luồng: {thread_name}")

def bai_6_1():
    threads = []
    for i in range(5):
        t = threading.Thread(target=worker_6_1, name=f"Luồng-{i+1}")
        threads.append(t)
        t.start()
    
    for t in threads:
        t.join()
```

```python
#6.2
def download_file(url, filename):
    print(f"Đang tải {filename} từ {url}...")
    try:
        time.sleep(2) 
        print(f"-> Hoàn tất tải {filename}")
    except Exception as e:
        print(f"Lỗi tải {filename}: {e}")

def bai_6_2():
    files_to_download = [
        ("http://example.com/file1.txt", "file1.txt"),
        ("http://example.com/file2.txt", "file2.txt"),
        ("http://example.com/file3.txt", "file3.txt")
    ]
    threads = []
    for url, name in files_to_download:
        t = threading.Thread(target=download_file, args=(url, name))
        threads.append(t)
        t.start()
    
    for t in threads:
        t.join()
```

```python
#6.3
def print_even():
    for i in range(30, 51):
        if i % 2 == 0:
            print(f"Even Thread: {i}")
            time.sleep(0.1)

def print_odd():
    for i in range(30, 51):
        if i % 2 != 0:
            print(f"Odd Thread: {i}")
            time.sleep(0.1)

def bai_6_3():
    print("\n--- BÀI 6.3 ---")
    t1 = threading.Thread(target=print_even)
    t2 = threading.Thread(target=print_odd)
    t1.start()
    t2.start()
    t1.join()
    t2.join()
```

```python
#6.4
factorial_results = []
def partial_factorial(start, end):
    result = 1
    for i in range(start, end + 1):
        result *= i
    factorial_results.append(result)

def bai_6_4():
    number = 10 
    mid = number // 2
    
    t1 = threading.Thread(target=partial_factorial, args=(1, mid))
    t2 = threading.Thread(target=partial_factorial, args=(mid + 1, number))
    t1.start()
    t2.start()
    t1.join()
    t2.join()
    final_result = 1
    for res in factorial_results:
        final_result *= res
    
    print(f"Giai thừa của {number} (tính bằng đa luồng) là: {final_result}")
    print(f"Kiểm tra bằng thư viện math: {math.factorial(number)}")
```

```python
#6.5
def make_request(url):
    print(f"Đang gửi request tới {url}")
    try:
        response = requests.get(url)
        print(f"Nhận phản hồi từ {url}: {response.status_code}")
    except Exception as e:
        print(f"Lỗi kết nối tới {url}")
print(make_request("http://example.com"))
```

```python
#6.5
def bai_6_5():
    urls = ["https://www.google.com", "https://www.python.org", "https://www.github.com"]
    threads = []
    for url in urls:
        t = threading.Thread(target=make_request, args=(url,))
        threads.append(t)
        t.start()
    
    for t in threads:
        t.join()
```

```python
#6.6
def write_evens(limit):
    for i in range(0, limit + 1, 2):
        print(f"Thread Chẵn: {i}")
        time.sleep(0.1)

def write_odds(limit):
    for i in range(1, limit + 1, 2):
        print(f"Thread Lẻ: {i}")
        time.sleep(0.1)

def bai_6_6():
    limit = 10
    t1 = threading.Thread(target=write_evens, args=(limit,))
    t2 = threading.Thread(target=write_odds, args=(limit,))
    t1.start()
    t2.start()
    t1.join()
    t2.join()
if __name__ == "__main__":
    bai_6_1()
    bai_6_2()
    bai_6_3()
    bai_6_4()
    bai_6_5()
    bai_6_6()
```

```python
#6.7
import datetime
class SocialThread(threading.Thread):
    def __init__(self, name, delay=1, counter=3):
        threading.Thread.__init__(self)
        self.name = name
        self.delay = delay
        self.counter = counter

    def run(self):
        print(f"Starting {self.name}")
        while self.counter > 0:
            time.sleep(self.delay)
            timestamp = datetime.datetime.now().strftime("Web %b %d %H:%M:%S %Y")
            print(f"{self.name}: {timestamp}")
            self.counter -= 1
        print(f"Exiting {self.name}")
def bai_6_7():
    print("Starting Main Thread")
    t1 = SocialThread("Google", delay=1, counter=4)
    t2 = SocialThread("Yahoo", delay=1, counter=4)
    t3 = SocialThread("Facebook", delay=1, counter=4)
    
    t1.start()
    t2.start()
    t3.start()
    
    t1.join()
    t2.join()
    t3.join()
    
    print("Exiting Main Thread")
```

```python
#6.8
class SumThread(threading.Thread):
    def __init__(self, data_list):
        threading.Thread.__init__(self)
        self.data_list = data_list
        self.result = 0 

    def run(self):
        self.result = sum(self.data_list)
        print(f"{self.name} đang tính tổng đoạn {self.data_list}. Kết quả: {self.result}")

def bai_6_8():
    n = 20 
    full_list = [random.randint(0, 10) for _ in range(n)]
    print(f"List gốc: {full_list}")
    num_threads = 4
    chunk_size = len(full_list) // num_threads
    threads = []
    for i in range(num_threads):
        start = i * chunk_size
        end = (i + 1) * chunk_size if i != num_threads - 1 else len(full_list)
        sub_list = full_list[start:end]
        
        t = SumThread(sub_list)
        threads.append(t)
        t.start()
    
    total_sum = 0
    for t in threads:
        t.join()
        total_sum += t.result
        
    print(f"Tổng tính bằng đa luồng: {total_sum}")
    print(f"Tổng kiểm tra (sum): {sum(full_list)}")
```

```python

class MaxThread(threading.Thread):
    def __init__(self, data_list):
        threading.Thread.__init__(self)
        self.data_list = data_list
        self.max_val = -1

    def run(self):
        if self.data_list:
            self.max_val = max(self.data_list)
        print(f"{self.name} tìm max trong {self.data_list}. Max cục bộ: {self.max_val}")

def bai_6_10():
    n = 20
    full_list = [random.randint(0, 100) for _ in range(n)]
    print(f"List gốc: {full_list}")
    
    num_threads = 4
    chunk_size = len(full_list) // num_threads
    threads = []
    
    for i in range(num_threads):
        start = i * chunk_size
        end = (i + 1) * chunk_size if i != num_threads - 1 else len(full_list)
        sub_list = full_list[start:end]
        
        t = MaxThread(sub_list)
        threads.append(t)
        t.start()
    
    global_max = -1
    for t in threads:
        t.join()
        if t.max_val > global_max:
            global_max = t.max_val
            
    print(f"Giá trị lớn nhất tìm được: {global_max}")
    print(f"Giá trị max kiểm tra: {max(full_list)}")

if __name__ == "__main__":
    bai_6_10()
```
