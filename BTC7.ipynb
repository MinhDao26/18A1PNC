```python
#7.5
import tkinter as tk
from PIL import Image, ImageTk 

def bai_7_5_show_image():
    root = tk.Tk()
    root.title("Bài 7.5: Hiển thị hình ảnh")
    root.geometry("400x450")
    tk.Label(root, text="CHƯƠNG TRÌNH HIỂN THỊ ẢNH", font=("Arial", 12, "bold")).pack(pady=10)

    try:
        load = Image.open("image.png") 
        load = load.resize((250, 250), Image.Resampling.LANCZOS)
        
        render = ImageTk.PhotoImage(load)
        img_label = tk.Label(root, image=render)
        img_label.image = render 
        img_label.pack(pady=10)
        
    except Exception as e:
        tk.Label(root, text=f"Lỗi: Không tìm thấy file 'image.png'\n({e})", fg="red").pack()
    btn_close = tk.Button(root, text="Đóng", command=root.destroy, width=10)
    btn_close.pack(pady=20)

    root.mainloop()
if __name__ == "__main__":
    bai_7_5_show_image()
```

```python
def bai_7_6():
    root = tk.Tk()
    root.title("Xử lý Chuỗi và Số")
    tk.Label(root, text="Enter a word:").grid(row=0, column=0)
    word_entry = tk.Entry(root)
    word_entry.grid(row=0, column=1)
    res_label = tk.Label(root, text="", fg="blue")
    res_label.grid(row=1, column=1)

    def reverse_word():
        text = word_entry.get()
        res_label.config(text=text[::-1])

    tk.Button(root, text="Validate (7.6)", command=reverse_word).grid(row=2, column=1)
    root.mainloop()
if __name__ == "__main__":
    bai_7_6()
```

```python
def bai_7_7():
    root = tk.Tk()
    root.title("Xử lý Chuỗi và Số")    
    tk.Label(root, text="Enter N:").grid(row=3, column=0)
    n_entry = tk.Entry(root)
    n_entry.grid(row=3, column=1)
    sum_label = tk.Label(root, text="")
    sum_label.grid(row=4, column=1)

    def calculate_sum():
        try:
            n = int(n_entry.get())
            total = sum(range(1, n + 1))
            sum_label.config(text=f"The sum is: {total}")
        except:
            sum_label.config(text="Vui lòng nhập số nguyên")

    tk.Button(root, text="Validate (7.7)", command=calculate_sum).grid(row=5, column=1)
    
    root.mainloop()
if __name__ == "__main__":
    bai_7_7()
```

```python
import tkinter as tk
from tkinter import ttk
import math

def bai_7_8_divisors():
    root = tk.Tk()
    root.title("Bài 7.8: Tìm ước số")
    
    tk.Label(root, text="Enter the value of N:").grid(row=0, column=0)
    ent_n = tk.Entry(root)
    ent_n.grid(row=0, column=1)
    
    lbl_res = tk.Label(root, text="The divisors of N: ")
    lbl_res.grid(row=1, column=0, columnspan=2)

    def find_divisors():
        try:
            n = int(ent_n.get())
            divs = [str(i) for i in range(1, n + 1) if n % i == 0]
            lbl_res.config(text=f"The divisors of N: {' '.join(divs)}")
        except:
            lbl_res.config(text="Vui lòng nhập số nguyên!")

    tk.Button(root, text="Validate", command=find_divisors).grid(row=2, column=1)
    root.mainloop()
if __name__ == "__main__":
    bai_7_8_divisors()
```

```python
def bai_7_9_gcd_lcm():
    root = tk.Tk()
    root.title("Bài 7.9: GCD & LCM")

    tk.Label(root, text="Enter value of m:").grid(row=0, column=0)
    ent_m = tk.Entry(root)
    ent_m.grid(row=0, column=1)

    tk.Label(root, text="Enter value of n:").grid(row=1, column=0)
    ent_n = tk.Entry(root)
    ent_n.grid(row=1, column=1)

    tk.Label(root, text="Choose function:").grid(row=2, column=0)
    combo = ttk.Combobox(root, values=["GCD", "LCM"])
    combo.grid(row=2, column=1)
    combo.current(0)

    lbl_res = tk.Label(root, text="Kết quả: ")
    lbl_res.grid(row=3, column=0, columnspan=2)

    def calculate():
        try:
            m, n = int(ent_m.get()), int(ent_n.get())
            if combo.get() == "GCD":
                res = math.gcd(m, n)
                lbl_res.config(text=f"GCD(m, n) = {res}")
            else:
                res = abs(m * n) // math.gcd(m, n)
                lbl_res.config(text=f"LCM(m, n) = {res}")
        except:
            lbl_res.config(text="Lỗi nhập liệu!")

    tk.Button(root, text="Calculate", command=calculate).grid(row=4, column=1)
    root.mainloop()
if __name__ == "__main__":
    bai_7_9_gcd_lcm()
```

```python
def bai_7_10_age_check():
    root = tk.Tk()
    root.withdraw() 
    
    from tkinter import simpledialog
    name = simpledialog.askstring("Input", "Nhập tên của bạn:")
    age = simpledialog.askinteger("Input", "Nhập tuổi của bạn:")
    
    if name and age is not None:
        msg = f"Xin chào {name}!\n"
        msg += "Bạn đã trưởng thành!" if age >= 18 else "Bạn còn nhỏ tuổi!"
        tk.messagebox.showinfo("Thông báo", msg)
    root.destroy()
if __name__ == "__main__":
    bai_7_10_age_check()
```

```python
def bai_7_11_lunar_year():
    root = tk.Tk()
    root.title("Đổi năm Dương lịch sang Âm lịch")

    tk.Label(root, text="Nhập năm Dương lịch:").pack()
    ent_year = tk.Entry(root)
    ent_year.pack()
    
    lbl_res = tk.Label(root, text="Năm Âm lịch: ", font=("Arial", 12, "bold"))
    lbl_res.pack()
    
    img_label = tk.Label(root) 
    img_label.pack()

    def convert():
        try:
            y = int(ent_year.get())
            can = ["Canh", "Tân", "Nhâm", "Quý", "Giáp", "Ất", "Bính", "Đinh", "Mậu", "Kỷ"]
            chi = ["Thân", "Dậu", "Tuất", "Hợi", "Tý", "Sửu", "Dần", "Mão", "Thìn", "Tỵ", "Ngọ", "Mùi"]
            
            ten_nam = f"{can[y % 10]} {chi[y % 12]}"
            lbl_res.config(text=f"Năm Âm lịch: {ten_nam}")
        except:
            lbl_res.config(text="Lỗi!")

    tk.Button(root, text="Chuyển đổi", command=convert).pack()
    root.mainloop()
if __name__ == "__main__":
    bai_7_11_lunar_year()
```

```python
def bai_7_12_bmi():
    root = tk.Tk()
    root.title("BMI Calculator")

    tk.Label(root, text="Cân nặng (kg):").pack()
    e_weight = tk.Entry(root)
    e_weight.pack()

    tk.Label(root, text="Chiều cao (m):").pack()
    e_height = tk.Entry(root)
    e_height.pack()

    lbl_result = tk.Label(root, text="BMI: ...")
    lbl_result.pack()
    lbl_status = tk.Label(root, text="Kết luận: ...")
    lbl_status.pack()

    def calc_bmi():
        try:
            w = float(e_weight.get())
            h = float(e_height.get())
            bmi = w / (h * h)
            lbl_result.config(text=f"BMI: {bmi:.2f}")
            
            if bmi < 18.5: status = "Gầy"
            elif bmi < 24.9: status = "Bình thường"
            else: status = "Thừa cân"
            lbl_status.config(text=f"Kết luận: {status}")
        except:
            messagebox.showerror("Lỗi", "Nhập sai định dạng số")

    tk.Button(root, text="Tính BMI", command=calc_bmi).pack()
    root.mainloop()
if __name__ == "__main__":
    bai_7_12_bmi()
```

```python
def bai_7_13_menu_system():
    root = tk.Tk()
    root.title("Hệ thống bài tập nâng cao Python")
    root.geometry("500x300")

    menubar = tk.Menu(root)

    def show_demo(bai_so):
        if bai_so == "1.1":
            tk.messagebox.showinfo("Lập trình hướng đối tượng", "Giải phương trình bậc 2")
        else:
            tk.messagebox.showinfo("Thông báo", f"Bạn đã chọn bài {bai_so}")
    for i in range(1, 5):
        chuong_menu = tk.Menu(menubar, tearoff=0)
        for j in range(1, 10):
            label = f"{i}.{j}"
            chuong_menu.add_command(label=label, command=lambda b=label: show_demo(b))
        
        if i == 1: chuong_menu.add_separator(); chuong_menu.add_command(label="Exit", command=root.quit)
        
        menubar.add_cascade(label=f"Chương {i}", menu=chuong_menu)

    root.config(menu=menubar)
    tk.Label(root, text="Chọn bài tập từ Menu phía trên", font=("Arial", 10)).pack(expand=True)
    root.mainloop()
if __name__ == "__main__":
    bai_7_13_menu_system()
```
